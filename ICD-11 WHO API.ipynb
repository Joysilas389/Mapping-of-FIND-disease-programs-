```python
# disable insecure request warning
requests.packages.urllib3.disable_warnings()

# set token endpoint and credentials
token_endpoint = 'https://icdaccessmanagement.who.int/connect/token'
client_id = '18e56bce-78d9-4b6f-a2c4-f471fc525936_fa22fa85-d247-4f76-b2ac-ea3be5fb3cb5'
client_secret = 'K04fmuPBZvWMOvZyzVyXLpzd9kV2MCPy4ojk2wYfE8A='
scope = 'icdapi_access'
grant_type = 'client_credentials'

# set data to post
payload = {'client_id': client_id, 
           'client_secret': client_secret, 
           'scope': scope, 
           'grant_type': grant_type}

# make request with certificate verification
r = requests.post(token_endpoint, data=payload, verify=True).json()
token = r['access_token']


# HTTP header fields to set
headers = {'Authorization':  'Bearer '+token, 
           'Accept': 'application/json', 
           'Accept-Language': 'en',
           'API-Version': 'v2'}

# access ICD API
uris = ['http://id.who.int/icd/release/11/2023-01/mms/257068234','http://id.who.int/icd/release/11/2023-01/mms/893358872'
       , 'http://id.who.int/icd/release/11/2023-01/mms/257068234', 'http://id.who.int/icd/release/11/2023-01/mms/257068234'
       , 'http://id.who.int/icd/release/11/2023-01/mms/257068234', 'http://id.who.int/icd/release/11/2023-01/mms/257068234'
       , 'http://id.who.int/icd/release/11/2023-01/mms/257068234', 'http://id.who.int/icd/release/11/2023-01/mms/257068234'
       , 'http://id.who.int/icd/release/11/2023-01/mms/257068234', 'http://id.who.int/icd/release/11/2023-01/mms/257068234'
       , 'http://id.who.int/icd/release/11/2023-01/mms/257068234', 'http://id.who.int/icd/release/11/2023-01/mms/257068234']

data_list = []

# loop over URIs and make request for each
for uri in uris:
    # make request with certificate verification
    r = requests.get(uri, headers=headers, verify=True)
    # convert JSON data into a Python data structure
    data = r.json()
    code = data.get('code', 'NaN')
    classKind = data['classKind']
    title = data['title']['@value']
    linearization_id = data['@id']
    # add data to list
    data_list.append([code, classKind, title, linearization_id])

df = pd.DataFrame(data_list, columns=['ICD 11 code', 'ClassKind', 'ICD 11 title', 'Linearization ID'])

# print dataframe
df
```
